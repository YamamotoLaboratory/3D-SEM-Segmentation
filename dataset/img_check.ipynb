```python
class Get_specfic_img(object):
    
    def __init__(self, path):
        
        self.dataset = self.__unzip(path)
        
    def __unzip(self, path):
        with gzip.GzipFile(path, "r") as f:
            data = np.load(f)
        return data
    
    def show_img(self, num, class_num=0):
        fig = plt.figure(figsize=(10, 5))
        
        sets = [self.dataset[0][num-1, :, :, 0], self.dataset[1][num-1, :, :, class_num]]
        
        ax = []
        for i in range(2):
            ax.append(fig.add_subplot(1, 2, i+1))
            ax[i].imshow(sets[i], cmap='gray')
        
        plt.show()
```

```python
path = './train_dataset/train_dataset.npy.gz'

obj = Get_specfic_img(path)

obj.show_img(2)
```
